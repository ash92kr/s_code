```python
# 영화별 상세정보 구하기

movie_name = []
time = []
year = []
nation = []
director = []
company = []
grade = []

movie_genre = []
genre1 = []
genre2 = []
genre3 = []

movie_actor = []
actor1 = []
actor2 = []
actor3 = []


for i in range(len(movie_code)):
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={0}&movieCd={1}'.format(private, movie_code[i])
    req = requests.get(url)
    movie = req.json()
#     movie = movie.decode('cp949').encode('utf-8')

#     print(movie)

    movie_name.append(movie["movieInfoResult"]["movieInfo"]["movieNm"])
    time.append(movie["movieInfoResult"]["movieInfo"]["showTm"])
    year.append(movie["movieInfoResult"]["movieInfo"]["openDt"])
#     director.append(movie["movieInfoResult"]["movieInfo"]["directors"][0]["peopleNm"])
    
    
    if len(movie["movieInfoResult"]["movieInfo"]["genres"]) >= 3:
        for j in range(3):
            movie_genre.append(movie["movieInfoResult"]["movieInfo"]["genres"][j]["genreNm"])
    elif len(movie["movieInfoResult"]["movieInfo"]["genres"]) == 2:
        for j in range(2):
            movie_genre.append(movie["movieInfoResult"]["movieInfo"]["genres"][j]["genreNm"])
        movie_genre.append('')
    elif len(movie["movieInfoResult"]["movieInfo"]["genres"]) == 1:
        movie_genre.append(movie["movieInfoResult"]["movieInfo"]["genres"][0]["genreNm"])
        movie_genre.append('')
        movie_genre.append('')
    else:
        movie_genre.append('')
        movie_genre.append('')
        movie_genre.append('')
    
    
    if len(movie["movieInfoResult"]["movieInfo"]["actors"]) >= 3:        
        for j in range(3):
            movie_actor.append(movie["movieInfoResult"]["movieInfo"]["actors"][j]["peopleNm"])
    elif len(movie["movieInfoResult"]["movieInfo"]["actors"]) == 2:
        for j in range(2):
            movie_actor.append(movie["movieInfoResult"]["movieInfo"]["actors"][j]["peopleNm"])
        movie_actor.append('')
    elif len(movie["movieInfoResult"]["movieInfo"]["actors"]) == 1:
        movie_actor.append(movie["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"])
        movie_actor.append('')
        movie_actor.append('')
    else:
        movie_actor.append('')
        movie_actor.append('')
        movie_actor.append('')
        
        
    if len(movie["movieInfoResult"]["movieInfo"]["nations"][0]["nationNm"]):
        nation.append(movie["movieInfoResult"]["movieInfo"]["nations"][0]["nationNm"])
    else:
        nation.append('')
    
    
    if len(movie["movieInfoResult"]["movieInfo"]["audits"]):
        grade.append(movie["movieInfoResult"]["movieInfo"]["audits"][0]["watchGradeNm"])
    else:
        grade.append('')
    

    if len(movie["movieInfoResult"]["movieInfo"]["companys"]):
        company.append(movie["movieInfoResult"]["movieInfo"]["companys"][0]["companyNm"])
    else:
        company.append('')
        
        
    if len(movie["movieInfoResult"]["movieInfo"]["directors"]):
        director.append(movie["movieInfoResult"]["movieInfo"]["directors"][0]["peopleNm"])
    else:
        director.append('')    
```

```python
def csv_writes(file_name, list1, list2, list3, list4, list5, list6, list7, list8, list9, list10, list11, list12, list13, list14):

    with open(file_name, 'w', newline='', encoding='utf-8') as f:
        fieldnames = ('movie_code', 'movie_name', 'time', 'year', 'nation', 'director', 'company', 'grade', 'genre1', 'genre2', 'genre3', 'actor1', 'actor2', 'actor3')
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()

        for i in range(len(list1)):
            writer.writerow({'movie_code': list1[i], 'movie_name': list2[i], 'time': list3[i], 'year': list4[i],
            'nation': list5[i], 'director': list6[i], 'company': list7[i], 'grade': list8[i], 'genre1': list9[i],
            'genre2': list10[i], 'genre3': list11[i], 'actor1': list12[i], 'actor2': list13[i], 'actor3': list14[i] })


csv_writes('movie.csv', movie_code, movie_name, time, year, nation, director, company,
           grade, genre1, genre2, genre3, actor1, actor2, actor3) 
```
