```python
import requests as req, os, datetime, json, csv
```

```python
def __get_file_column_datas(filename, column_name):
    movie_code_list = []
    with open(filename, 'r', newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            movie_code_list.append(row[column_name])
    return movie_code_list
```

```python
def __write_movie_file(filename, fieldnames, datas):
    with open(filename, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for data in datas:
            writer.writerow(data)
```

```python
movie = []
movie_name = __get_file_column_datas('movie_naver_excel.csv', 'movie_name')
time = __get_file_column_datas('movie_naver_excel.csv', 'time')
year = __get_file_column_datas('movie_naver_excel.csv', 'year')
nation = __get_file_column_datas('movie_naver_excel.csv', 'nation')
director = __get_file_column_datas('movie_naver_excel.csv', 'director')
company = __get_file_column_datas('movie_naver_excel.csv', 'company')
grade = __get_file_column_datas('movie_naver_excel.csv', 'grade')
genre = genre_list[1:]
actor1 = actor1_list[1:]
actor2 = actor2_list[1:]
actor3 = actor3_list[1:]
poster_url = __get_file_column_datas('movie_naver_excel.csv', 'poster_url')
link_url = __get_file_column_datas('movie_naver_excel.csv', 'link_url')
user_rating = __get_file_column_datas('movie_naver_excel.csv', 'user_rating')
datas = list(zip(movie_name,time,year,nation,director,company,grade,genre,actor1,actor2,actor3,poster_url,link_url,user_rating))

for data in datas:        
    save_data = {
        'movie_name': data[0],
        'time': data[1],
        'year': data[2],
        'nation': data[3],
        'director': data[4],
        'company': data[5],
        'grade': data[6],
        'genre': data[7],
        'actor1': data[8],
        'actor2': data[9],
        'actor3': data[10],
        'poster_url': data[11],
        'link_url': data[12],
        'user_rating': data[13]
    }
    movie.append(save_data)
__write_movie_file('movie_not_list.csv', save_data.keys(), movie)
```
